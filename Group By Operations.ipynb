```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
#maximum vs grouped maximum on series 
#Group by
df['Weight'].max()
```

```python
#maximum weight for various groups by origin
df['Weight'].groupby(df['Origin']).max()
```

```python
#group by from dataframe
#group by origin in DataFrames
df.groupby('Origin').max()
```

```python
#group by from DataFrames and Filter Data
df.groupby('Origin').max()['Weight']
```

```python
df.groupby('Origin')['Weight'].max()
```

```python
#group by multiples series
df.groupby(['Origin', 'Cylinders']).max()
```

```python
#group by multiples series and Filter series 
df.groupby(['Origin', 'Cylinders']).max()[['Weight', 'Model']]
```

```python
#group by custom aggregation
df.groupby('Origin').agg({
    'Car':'first',
    'Cylinders':'max',
    'Acceleration':'mean',
    'Weight':'max',
    'Model':'min'
})
```
