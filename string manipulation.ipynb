```python
#Import Library and Data Load
import pandas as pd
```

```python
#Load Dataset
df = pd.read_csv('cars.csv')
df
```

```python
#To upper and lower case
#To upper case
df['Car'] = df['Car'].str.upper()
df.head()
```

```python
#To lower case
df['Car'] = df['Car'].str.lower()
df.head()
```

```python
#to capitalize
df['Car'] = df['Car'].str.capitalize()
df.head()
```

```python
#Replace a specific string
df['Car'] = df['Car'].str.replace('Ford','My')
df.head()
```

```python
#starts with and ends with
#starts with
condition = df['Car'].str.startswith('My')
df.loc[condition].head()
```

```python
#ends with 
condition = df['Car'].str.endswith('torino')
df.loc[condition].head()
```

```python
#string contains
condition = df['Car'].str.contains('torino')
df.loc[condition].head()
```

```python
#split string
df['Car'].str.split()
```
