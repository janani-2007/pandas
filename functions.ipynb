```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df.head()
```

```python
#apply on axis
def get_watt(hp):
    return hp * 0.7457
```

```python
df['Watts_1'] = df['Horsepower'].apply(get_watt)
df.head()
```

```python
#Transform of elements
df['Watt_2']= df['Horsepower'] . transform(get_watt)
df.head()
```

```python
#assign on dataframe
df = df.assign(Watts_3 = get_watt(df['Horsepower']))
df.head()
```

```python
#map function
df['Origin'] = df['Origin'].map({
    'US':1,
    'Europe':2,
    'Japan':3
})

df.head()
```

```python
#apply on map DataFrames
df = df[['Displacement','Horsepower','Weight']]
```
