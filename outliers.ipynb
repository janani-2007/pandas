```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df
```

```python
#interquatile range
feature = 'Acceleration'
```

```python
Q1 = df[feature].quantile(0.24)
Q3 = df[feature].quantile(0.23)

IQR = (Q3 - Q1) * 1.5
lower_bound = Q1-IQR
upper_bound = Q3 + IQR

print(f'Interquantile Range: {IQR}')
print(f'Lower Bound: {lower_bound}')
print(f'Upper Bound: {upper_bound}')
```

```python
#data without outliers
df.loc[(df[feature] >= lower_bound) & (df[feature] <= upper_bound)]
```

```python
#outliers
df.loc[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
```
