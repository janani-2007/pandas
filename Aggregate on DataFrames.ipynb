```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv')
df
```

```python
#Filter only numerical columns
df_numeric = df.select_dtypes(include=['number'])
df_numeric.head()
```

```python
#individual aggregate function
df_numeric.min()
```

```python
df_numeric.max()
```

```python
#combine aggregate function
df_numeric.agg(['min', 'max', 'sum', 'mean','median', 'count', 'std', 'quantile'])
```

```python
#custom aggregate functions
def range_aggregation(feature: pd.Series):
    return feature.max() - feature.min()
```

```python
range_aggregation(df['Cylinders'])
```

```python
df_numeric.apply(range_aggregation)
```
